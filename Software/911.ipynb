```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./911-all - 911.csv')
```

```python
df.head()
```

```python
# Filtrar los datos más recientes (2022-2023) usando el vector de municipios
municipios_vector = [39]  # Agrega los municipios que desees filtrar
filtered_df = df[(df['X.periodo'] >= 'p2022-2023') & (df['cve.municipio'].isin(municipios_vector))]
```

```python
filtered_df
```

```python
# Filtrar las carreras tecnológicas que comienzan con 54, 55, 56 y 57
carreras_tecnologicas = filtered_df[filtered_df['programa'] == 'TECNOLOGICO']
```

```python
carreras_tecnologicas
```

```python
total_carreras_tecnologicas = carreras_tecnologicas['nuevo.ingreso'].sum() + carreras_tecnologicas['egresados'].sum() + carreras_tecnologicas['titulados'].sum()
print("Total de carreras tecnológicas (2022-2023):", total_carreras_tecnologicas)
```

```python
historico_carreras_tecnologicas = df[(df['X.periodo'] >= '2014') & df['carrera'].str.startswith(('54', '55', '56', '57'))]
```

```python
# Mostrar los resultados
print("Datos filtrados:")
print(filtered_df)

print("\nTotal de estudiantes en carreras tecnológicas (2022-2023):", total_carreras_tecnologicas)

print("\nHistórico de estudiantes desde 2014-2015:")
print(historico_carreras_tecnologicas)
```
